```python
#2. 

class Stack:
    def __init__(self):
        self.stack = []

    def makeEmpty(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()
        else:
            return None

    def isEmpty(self):
        return len(self.stack) == 0

s= Stack()
x = 3
y = 1
s.makeEmpty()    #Khởi tạo một ngăn xếp trống.
s.push(5)        #Đẩy giá trị 5 vào ngăn xếp.
s.push(7)        #Đẩy giá trị 7 vào ngăn xếp.
s.pop()          #Lấy phần tử trên cùng ra khỏi ngăn xếp (7 bị xóa).
x += y           #x được cập nhật thành 4 (3 + 1).
s.pop()          #Lấy phần tử trên cùng ra khỏi ngăn xếp (5 bị xóa).
s.push(x)        #Đẩy giá trị của x (4) vào ngăn xếp.
s.push(y)        #Đẩy giá trị của y (1) vào ngăn xếp.
s.push(2)        #Đẩy giá trị 2 vào ngăn xếp.
s.pop()          #Lấy phần tử trên cùng ra khỏi ngăn xếp (2 bị xóa).
s.pop()          #Lấy phần tử trên cùng ra khỏi ngăn xếp (1 bị xóa).

#Vòng lặp while bật và in các phần tử cho đến khi ngăn xếp trống.
while not s.isEmpty():
    y = s.pop()
    print(y)
print('x =', x)
print('y =', y)
```

```python
#Question 5:

from collections import deque

q = deque()
sequence = ['E', 'A', 'S', '*', 'Y', '*', 'Q', 'U', 'E', '*', '*', '*', 'S', 'T', '*', '*', '*', 'I', 'O', '*', 'N', '*', '*', '*']
results = []

for item in sequence:
    if item != '*':
        q.append(item)
    else:
        results.append(q.popleft())
```

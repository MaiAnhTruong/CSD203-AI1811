```python

class EmptyStackException(Exception):
    pass

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stack:
    def __init__(self):
        self.top_node = None

    def isEmpty(self):
        return self.top_node is None

    def clear(self):
        self.top_node = None

    def push(self, x):
        new_node = Node(x)
        new_node.next = self.top_node
        self.top_node = new_node

    def pop(self):
        if self.isEmpty():
            raise EmptyStackException("Ngăn xếp rỗng")
        value = self.top_node.value
        self.top_node = self.top_node.next
        return value

    def top(self):
        if self.isEmpty():
            raise EmptyStackException("Ngăn xếp rỗng")
        return self.top_node.value

    def traverse(self):
        current = self.top_node
        while current:
            print(current.value, end=" ")
            current = current.next
        print()

def decimal_to_binary(decimal_number):
    stack = Stack()
    while decimal_number > 0:
        remainder = decimal_number % 2
        stack.push(remainder)
        decimal_number //= 2

    binary_representation = ""
    while not stack.isEmpty():
        binary_representation += str(stack.pop())

    return binary_representation

# Ví dụ sử dụng:
ngan_xep_vd = Stack()
ngan_xep_vd.push(1)
ngan_xep_vd.push(2)
ngan_xep_vd.push(3)

print("Ngăn xếp có rỗng không?", ngan_xep_vd.isEmpty())  
print("Phần tử đầu ngăn xếp:", ngan_xep_vd.top())        
ngan_xep_vd.traverse()                                 

print("Phần tử đã rút ra:", ngan_xep_vd.pop())          
ngan_xep_vd.traverse()                                 

ngan_xep_vd.clear()
print("Ngăn xếp có rỗng sau khi xóa?", ngan_xep_vd.isEmpty()) 

so_thap_phan = 25
bieu_dien_nhi_phan = decimal_to_binary(so_thap_phan)
print(f"Bản biểu diễn nhị phân của {so_thap_phan}: {bieu_dien_nhi_phan}")
```
